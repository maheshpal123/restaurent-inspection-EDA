```python


# Convert the 'Violations' column to a string type
df1['Violations'] = df1['Violations'].astype(str)

# Define a function that takes a string and returns the count of comma-separated values
def count_values(s):
    
    if pd.isna(s):
        
        return 0
    else:
        
        return len(s.split(','))

# Apply the function to each row of the 'Violations' column and create a new column with the result
df1['Violation Count'] = df1['Violations'].apply(count_values)
```

## Q 4- Is Current Demerits a deciding factor for the Inspection results or Inspection Demerits?


```python
import pandas as pd


# Select only the columns of interest
selected_cols = ['Current Demerits', 'Inspection Demerits']
df = df[selected_cols]

# Calculate the correlation coefficient between the two columns
corr_coef = df.corr().iloc[0,1]

# Print the correlation coefficient
print(f'The correlation coefficient between Current Demerits and Inspection Demerits is: {corr_coef}')
```

```python
'''If the correlation coefficient is close to 1, it suggests that there is a strong positive correlation between
"Current Demerits" and "Inspection Demerits", meaning that as the number of current demerits increases,
the number of inspection demerits also tends to increase. If the correlation coefficient is close to -1,
it suggests that there is a strong negative correlation between the two variables, meaning that as
the number of current demerits increases, the number of inspection demerits tends to decrease. 
If the correlation coefficient is close to 0, it suggests that there is little or no correlation between the two variables.'''
```

## Q 5-The agency requires a visual representation of the distribution of the restaurants City vise who are under the ‘Re-inspection’ category of the Inspection type who still failed 
## the norms of the inspection with ‘Closed with Fees’, ‘Closed without Fees’ and ‘Not 
## Approved’ Inspection Results?
# 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the data to get restaurants that meet the conditions
condition = (df2['Inspection Type'] == 'Re-inspection') & \
            (df2['Inspection Result'].isin(['Closed with Fees', 'Closed without Fees', 'Not Approved']))
filtered_data = df2[condition]

# Group the data by city and count the number of restaurants
grouped_data = filtered_data.groupby('City')['Category Name'].count().reset_index()

# Sort the data in descending order of the number of restaurants
sorted_data = grouped_data.sort_values('Category Name', ascending=False)

# Create the bar plot using seaborn
sns.set_style('darkgrid')
plt.figure(figsize=(12, 8))
sns.barplot(x='City', y='Category Name', data=sorted_data)
plt.title('Distribution of restaurants by City under Re-inspection category')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

## Q 6-Visualize the distribution of the restaurants Category Type vise which have more than 
## 10 number violations and find the top 2 category types. Also, find the category type 
## with highest overall violations for all restaurants?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter dataframe for restaurants with more than 10 violations
df_filtered = df2[df2['Violation Count'] > 10]

# Group by Category Type and count the number of restaurants in each category
category_counts = df_filtered.groupby('Category Name').size().reset_index(name='Counts')

# Visualize the distribution of restaurants Category Name wise
plt.figure(figsize=(20, 10))
plt.bar(category_counts['Category Name'], category_counts['Counts'])
plt.xticks(rotation=90)
plt.xlabel('Category Name')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Restaurants with more than 10 violations by Category Name')

# Sort the groups by count in descending order and select the top 2 Category Types
top_categories = category_counts.sort_values(by='Counts', ascending=False).head(2)
print('Top 2 Category Names with more than 10 violations:')
print(top_categories['Category Name'].values)

# Find the sum of violations number for each Category Type using the original dataframe
violation_sums = df2.groupby('Category Name')['Violation Count'].sum().reset_index(name='Violation Sum')

# Sort the sums in descending order and select the Category Type with the highest overall violations
highest_violation_category = violation_sums.sort_values(by='Violation Sum', ascending=False).iloc[0]['Category Name']
print('Category Name with the highest overall violations:')
print(highest_violation_category)
```
